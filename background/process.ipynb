```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ...existing code...
we = cv2.imread("we.png")
fl = cv2.imread("flamingo.jpg")

we_h, we_w = we.shape[:2]
fl_h, fl_w = fl.shape[:2]

# 计算缩放后的 flamingo 高度
new_w = we_w
scale = new_w / fl_w
new_h = int(fl_h * scale)

# 缩放 flamingo
fl_resized = cv2.resize(fl, (new_w, new_h))

# 创建白色背景
result = 255 * np.ones((we_h, we_w, 3), dtype=np.uint8)

# 计算上下留白
y_offset = (we_h - new_h) // 2

# 将 flamingo 粘贴到白色背景中央
result[y_offset:y_offset+new_h, 0:new_w] = fl_resized

# 保存或显示结果
cv2.imwrite("flamingo_resized.png", result)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
